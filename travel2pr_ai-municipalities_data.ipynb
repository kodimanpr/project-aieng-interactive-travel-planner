```python
import os
import re
from collections import Counter

# Define the directory containing the municipality text files
data_path = r"C:\\Users\\larry\\OneDrive\\Documents\\GitHub\\project-aieng-interactive-travel-planner\\data\\municipalities"

# Initialize a Counter to store occurrences of special characters
special_chars_counter = Counter()

# Function to extract potential special characters from text
def extract_special_chars(text):
    # Match escaped sequences like \xc3\xa1, \xe2\x80\x99, etc.
    special_chars = re.findall(r'\\x[a-fA-F0-9]{2}\\x[a-fA-F0-9]{2}', text)
    return special_chars

# Scan all text files in the directory
for file in os.listdir(data_path):
    if file.endswith(".txt"):
        file_path = os.path.join(data_path, file)
        
        with open(file_path, "r", encoding="utf-8") as f:
            content = f.read()
            found_chars = extract_special_chars(content)
            special_chars_counter.update(found_chars)

# Convert the results to a sorted list of the most common characters
special_chars_list = special_chars_counter.most_common()

# Display results
import pandas as pd
df_special_chars = pd.DataFrame(special_chars_list, columns=["Encoded Character", "Frequency"])
#import ace_tools as tools
#tools.display_dataframe_to_user(name="Special Characters Frequency", dataframe=df_special_chars)
```

```python
import os
import json
import re
import requests
import unicodedata
import ftfy
from bs4 import BeautifulSoup

# Define the directories
#data_path = r"C:\\Users\\larry\\OneDrive\\Documents\\GitHub\\project-aieng-interactive-travel-planner\\data\\municipalities"
#output_path = r"C:\\Users\\larry\\OneDrive\\Documents\\GitHub\\project-aieng-interactive-travel-planner\\data\\municipalities_cleaning\\step1_cleaned_data"

data_path = r"C:\\Users\\larry\\OneDrive\\Documents\\GitHub\\project-aieng-interactive-travel-planner\\data\\municipalities\\test"
output_path = r"C:\\Users\\larry\\OneDrive\\Documents\\GitHub\\project-aieng-interactive-travel-planner\\data\\municipalities_cleaning\\\08feb25_test03\\html_spaces_geo_clean"

# Create output directory if it doesn't exist
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Step 1: Function for Fixing Special Characters
def fix_special_characters(text):
    text = unicodedata.normalize("NFKC", text)  # Normalize Unicode text
    text = ftfy.fix_text(text)  # Fix corrupted encodings
    return text

# Step 2: Function to extract geo-metadata from text
def extract_coordinates(text):
    coordinate_pattern = re.compile(r'(-?\d+\.\d+);\s*(-?\d+\.\d+)')  # Match pattern with semicolon separator
    matches = coordinate_pattern.findall(text)
    if matches:
        lat, lon = float(matches[0][0]), float(matches[0][1])  # Correct order [latitude, longitude]
        return [lat, lon]
    return None

# Step 3: Clean Function
def clean_text(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        raw_content = f.read()
    
    # Parse with BeautifulSoup
    soup = BeautifulSoup(raw_content, "html.parser")
    
    # Insert space after <a> tags and remove duplicates
    for a in soup.find_all("a"):
        a.insert_after(" ")
    
    # Extract text from relevant tags
    paragraphs = soup.find_all(['h1', 'h2', 'p'])
    cleaned_text = "\n".join([para.get_text(" ", strip=True) for para in paragraphs])
    
    # Apply character replacement
    cleaned_text = fix_special_characters(cleaned_text)
    
    # Extract coordinates
    coordinates = extract_coordinates(cleaned_text)
    
    # Extract municipality name
    municipality_name = os.path.basename(file_path).replace(".txt", "")
    
    # Construct geo-metadata
    geo_metadata = {
        "coordinates": coordinates if coordinates else "Not Found",
        "title": municipality_name,
        "type": "Point"
    }
    
    # Combine cleaned data
    return {
        "main_content": cleaned_text,
        "geo_metadata": geo_metadata
    }

# Apply to All Files
for file in os.listdir(data_path):
    file_path = os.path.join(data_path, file)
    cleaned_data = clean_text(file_path)
    output_file_path = os.path.join(output_path, file)

    with open(output_file_path, "w", encoding="utf-8") as out_file:
        json.dump(cleaned_data, out_file, indent=4, ensure_ascii=False)

    print(f"Processed: {file}")
```

```python
import os
import json
import requests

# Define directories
cleaned_data_path = r"C:\\Users\\larry\\OneDrive\\Documents\\GitHub\\project-aieng-interactive-travel-planner\\data\\landmarks_cleanng\\step1_cleaned_data"#cleaned_data"
final_data_path = r"C:\\Users\\larry\\OneDrive\\Documents\\GitHub\\project-aieng-interactive-travel-planner\\data\\landmarks_cleanng\\step2_cleaned_data"#final_data"

# Create the final output directory if it doesn't exist
if not os.path.exists(final_data_path):
    os.makedirs(final_data_path)

# Function to validate and deduplicate links
def process_relevant_links(links):
    """
    Validates and deduplicates a list of relevant links.
    Only keeps unique, valid links that start with "http".
    """
    # Filter for HTTP(S) links
    http_links = [link for link in links if link.startswith("http")]
    
    # Deduplicate links
    unique_links = list(set(http_links))
    
    # Validate links
    valid_links = []
    for link in unique_links:
        try:
            response = requests.head(link, timeout=5)
            if response.status_code == 200:
                valid_links.append(link)
        except requests.RequestException:
            continue

    return valid_links

# Process each file in the cleaned_data directory
for file_name in os.listdir(cleaned_data_path):
    file_path = os.path.join(cleaned_data_path, file_name)
    with open(file_path, "r", encoding="utf-8") as file:
        data = json.load(file)

    # Process relevant links if they exist
    if "relevant_links" in data:
        data["relevant_links"] = process_relevant_links(data["relevant_links"])

    # Save the updated data into the final_data directory
    final_file_path = os.path.join(final_data_path, file_name)
    with open(final_file_path, "w", encoding="utf-8") as final_file:
        json.dump(data, final_file, indent=4, ensure_ascii=False)

    print(f"Processed and saved: {file_name}")
```

```python
import os
import unicodedata
import ftfy

def normalize_text(text):
    # Normaliza el texto a su forma NFKD y filtra los caracteres ASCII
    return ''.join(c for c in unicodedata.normalize('NFKD', text) if unicodedata.category(c) != 'Mn')

def fix_text(text):
    return ftfy.fix_text(text)

data_path = r"C:\Users\larry\OneDrive\Documents\GitHub\project-aieng-interactive-travel-planner\data\municipalities\test"
output_path = r"C:\Users\larry\OneDrive\Documents\GitHub\project-aieng-interactive-travel-planner\data\municipalities_cleaning\08feb25_test01\special_characters_test"

# Crear el directorio de salida si no existe
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Procesar cada archivo en el directorio
for file in os.listdir(data_path):
    file_path = os.path.join(data_path, file)
    
    with open(file_path, "r", encoding="utf-8") as f:
        raw_content = f.read()
    
    # Aplicar normalización
    normalized_text = normalize_text(raw_content)
    
    # Aplicar corrección
    cleaned_text = fix_text(normalized_text)
    
    # Guardar el archivo corregido
    output_file_path = os.path.join(output_path, file)
    with open(output_file_path, "w", encoding="utf-8") as out_file:
        out_file.write(cleaned_text)

    print(f"Processed: {file}")

print("Normalization and Text Fixing Complete for All Files")
```

# **Municipalities Data Cleaning Step 3: Removal of pronunciation sections and addition of related en.wikipedia.org to the relevant links**

```python
import os
import json
import re

# Define input and output directories
final_data_path = r"C:\\Users\\larry\\OneDrive\\Documents\\GitHub\\project-aieng-interactive-travel-planner\\data\\municipalities_cleaning\\step2_relevant_links_ref_num"
final_cleaned_data_path = r"C:\\Users\\larry\\OneDrive\\Documents\\GitHub\\project-aieng-interactive-travel-planner\\data\\municipalities_cleaning\\step3_final_cleaned_data"

# Create output directory if it doesn't exist
if not os.path.exists(final_cleaned_data_path):
    os.makedirs(final_cleaned_data_path)

# Function to remove pronunciation sections (generic patterns within parentheses)
def remove_pronunciation(text):
    # Define the regex pattern to match different pronunciation formats
    pronunciation_pattern = re.compile(
        r"\(\s*(US:.*?|UK:.*?|Spanish:.*?|US :.*?|UK :.*?|Spanish :.*?|.*?pronunciation:.*?|/.*?/.*?)\s*\)",  # Include "pronunciation:" explicitly
        flags=re.DOTALL
    )
    # Remove all matching patterns
    cleaned_text = re.sub(pronunciation_pattern, "", text)
    return cleaned_text

# Function to ensure the relevant Wikipedia link is present
def ensure_wikipedia_link(links, municipality_name):
    # Generate the expected Wikipedia link
    formatted_name = municipality_name.replace(" ", "_")
    wikipedia_link = f"https://en.wikipedia.org/wiki/{formatted_name},_Puerto_Rico"
    
    # Check if the link is already in the list, if not, add it
    if wikipedia_link not in links:
        links.append(wikipedia_link)
    
    return links

# Process each file in the final_data directory
for file_name in os.listdir(final_data_path):
    file_path = os.path.join(final_data_path, file_name)

    with open(file_path, "r", encoding="utf-8") as file:
        data = json.load(file)

    # Clean main content to remove pronunciation sections
    if "main_content" in data:
        data["main_content"] = remove_pronunciation(data["main_content"])

    # Ensure relevant links contain the Wikipedia link for the municipality
    if "relevant_links" in data:
        municipality_name = os.path.splitext(file_name)[0]  # Extract municipality name from the file name
        data["relevant_links"] = ensure_wikipedia_link(data["relevant_links"], municipality_name)

    # Save the cleaned file to the final_cleaned_data directory
    output_file_path = os.path.join(final_cleaned_data_path, file_name)
    with open(output_file_path, "w", encoding="utf-8") as output_file:
        json.dump(data, output_file, indent=4, ensure_ascii=False)

    print(f"Processed and saved: {file_name}")
```

```python
import os

def generate_municipality_mapping(directory_path):
    """
    Reads all .txt files in the specified directory and generates
    a dictionary mapping municipality names to "Municipality, PR".
    """
    municipality_mapping = {}

    # Get all .txt files in the directory
    txt_files = [f for f in os.listdir(directory_path) if f.endswith(".txt")]

    for file in txt_files:
        municipality_name = os.path.splitext(file)[0]  # Remove ".txt" extension
        formatted_name = f"{municipality_name}, PR"  # Format for weather API
        municipality_mapping[municipality_name] = formatted_name

    return municipality_mapping  # Ensure the function returns the dictionary

# Example usage
directory_path = r"C:\Users\larry\OneDrive\Documents\GitHub\project-aieng-interactive-travel-planner\data\municipalities_cleaning\step3_final_cleaned_data"

# Store the returned dictionary in a variable
municipality_mapping = generate_municipality_mapping(directory_path)

# Save the mapping to a Python file
output_file = "municipality_mapping.py"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(f"municipality_mapping = {municipality_mapping}\n")

print(f"Municipality mapping saved to {output_file}")
```
